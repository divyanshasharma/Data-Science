### 1. WAP to demonstrate graphical display of data distribution using:
###    box plot, bar plot, area plot.

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
import pylab as py
import scipy.stats as stats
```

```python
np.random.seed(5)
data_1 = np.random.normal(100, 10, 200) 
data_2 = np.random.normal(90, 20, 200) 
data_3 = np.random.normal(80, 30, 200) 
data_4 = np.random.normal(70, 40, 200) 
data = [data_1, data_2, data_3, data_4] 
print("Data is: ",data)
```

```python
fig=plt.figure(figsize=(10,10))
plt.boxplot(data)
plt.show()
```

```python
plt.bar(data_1, data_2)
plt.show()
```

```python
df=pd.DataFrame({
    'data_1':data_1,
    'data_2':data_2,
})
ax=df.plot.area()
```

## 2. WAP for visual inspection of your data using:
### Quantile-Quantile plot (QQ-Plot), histogram, and scatter plot.

```python
data_1_normalized=(data_1-data_1.mean())/data_1.std(ddof=0)
sm.qqplot(data_1_normalized,line='45')
py.show()
```

```python
plt.hist(data_1)
plt.show()
```

```python
plt.scatter(data_1,data_2)
plt.show()
```

### 3. Suppose that a hospital tested the age and body fat data for 18 randomly selected adults with the following results:

    age  23   23  27  27  39  41  47  49  50 52  54   54  56  57  58  58   60  61
    %fat 9.5  26.5 7.8 17.8 31.4  25.9 27.4 27.2 31.2 34.6 42.5 28.8 33.4 30.2 34.1 32.9  41.2  35.7

   
    

### a) Draw the boxplots for age and %fat.

```python
age =[ 23   ,23,    27,   27 ,   39   , 41   , 47 ,   49,    50,52,    54,    54,    56,    57,    58,    58,    60,    61]
fat=[ 9.5  ,26.5 , 7.8  ,17.8  ,31.4  ,25.9  ,27.4  ,27.2  ,31.2,34.6, 42.5,  28.8,  33.4,  30.2,  34.1,  32.9,  41.2, 35.7]
data=[age,fat]
fig=plt.figure(figsize=(10,10))
plt.boxplot(data)
plt.show()
```

### (b) Draw a scatter plot and a q-q plot based on these two variables.

```python
plt.scatter(age,fat)
plt.show()
```

```python
df=pd.DataFrame({
    'Age':age,
    'Fat':fat,
})
ax=df.plot.area()
```

```python
n_rows=df.shape[0]
df['count']=df.index+1;
df['percentile_area']=(df['count'])/n_rows
df['z_Age']=(df['Age']-df['Age'].mean())/df['Age'].std(ddof=1)
df['z_Fat']=(df['Fat']-df['Fat'].mean())/df['Fat'].std(ddof=1)
```

```python
print(df)
```

```python
plt.scatter(df.z_Age,df.z_Fat)
plt.show()
```

```python
sm.qqplot(df.z_Age,line='45')
py.show()
```

```python
sm.qqplot(df.z_Fat,line='45')
py.show()
```
