### ID3
ID3 algorithm, stands for Iterative Dichotomiser 3, is a classification algorithm that follows a greedy approach of building a decision tree by selecting a best attribute that yields maximum Information Gain (IG) or minimum Entropy (H).


```python
import numpy as np
import pandas as pd
```

```python
data = [
    [ 1 , 'NO' , 'NO' , 'NO' , 'NO' ],

[ 2 , 'YES' , 'YES' , 'YES' , 'YES' ],

[ 3 , 'YES' , 'YES' , 'NO' , 'NO' ],

[ 4 , 'YES' , 'NO' , 'YES' , 'YES'],

[ 5 , 'YES' , 'YES' , 'YES' , 'YES' ],

[ 6 , 'NO' , 'YES' , 'NO' , 'NO' ],

[ 7 , 'YES' , 'NO' , 'YES', 'YES' ],

[ 8 , 'YES' , 'NO', 'YES' , 'YES' ],

[ 9 ,'NO' , 'YES' , 'YES' , 'YES'],

[ 10 , 'YES' , 'YES' , 'NO' , 'YES' ],

[ 11 , 'NO' , 'YES' , 'NO' , 'NO' ],

[ 12 , 'NO' , 'YES' , 'YES' , 'YES' ],

[ 13 , 'NO' , 'YES' , 'YES' , 'NO' ],

[ 14 , 'YES' , 'YES' , 'NO' , 'NO'   ]]

# Create the pandas DataFrame 
dff = pd.DataFrame(data, columns = ['ID' , 'Fever' , 'Cough' , 'Breathing issues' , 'Infected' ]) 
  
# print dataframe. 
dff
```

```python
import matplotlib.pyplot as plt

import sklearn
from sklearn import preprocessing
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


```

```python
labelEnc=preprocessing.LabelEncoder()
dff['Fever']=labelEnc.fit_transform(dff['Fever'])
dff['Cough']=labelEnc.fit_transform(dff['Cough'])
dff['Breathing issues']=labelEnc.fit_transform(dff['Breathing issues'])
dff['Infected']=labelEnc.fit_transform(dff['Infected'])
dff
```

```python
x=dff.iloc[:,0:4]
y=dff.loc[:,'Infected']
x
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x_train
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=2)
fig=model.fit(x_train,y_train)
```

```python
PreModel=model.predict(x_test)
print(round(accuracy_score(y_test,PreModel)))
```

```python
tree.plot_tree(fig)
plt.show()
```

download data from https://www.kaggle.com/uciml/pima-indians-diabetes-database 

```python
df=pd.read_csv("diabetes.csv")
df
```

```python
#Checking for missing data
Null_val=df.columns[df.isnull().any()]
df[Null_val].isnull().sum()
```

```python
x=df.iloc[:,0:8]
y=df.loc[:,'Outcome']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
x_train.shape
```

```python
DT=DecisionTreeClassifier(max_depth=3)
DT.fit(x_train,y_train)
```

```python
PreDT=DT.predict(x_test)
print(round(accuracy_score(y_test,PreDT)*100,2))
```
